## Getting AKR frequency limits
In the next two steps we convert the burst list into a file that is a time series of all sweeps in Waters mask with the AKR frequency limits from the burst list along with sweep number and burst number

First we convert the burst list into a time series of frequency limits

```python
from Wind_Waves.integration_tools import get_akr_flims
from Wind_Waves.reading_tools import find_closest
import numpy as np
import pandas as pd
fogg= get_akr_flims('../Example_Data/Full_Run_Through/fogg_akr_burst_list_1995_2004.csv')
waters= pd.read_hdf('../Example_Data/Full_Run_Through/Waters_mask.hdf5', key='main')
# There is a small mismatch between datetimes but most mismatches are small
new_dates= find_closest(fogg.Date_UTC.values, waters.Date_UTC.values)
# We remove cases where the mismatches are greater than one minute in our runs this only occurs 4 times
rm_ind= (new_dates-fogg.Date_UTC.values)>np.timedelta64(1, 'm')
fogg= fogg[~rm_ind]
fogg['Date_UTC']= new_dates[~rm_ind]
```

Next we join the burst list frequency limits to the full AKR mask dataset and reduce down to one data point per sweep and save the file

```python
df=waters.merge(fogg, on='Date_UTC', how='left')
df.loc[df.Burst_Number.isnull(), ['Burst_Number']]=-1
df.drop_duplicates(subset=['SWEEP'], keep='first', inplace=False)\
    [['Date_UTC', 'SWEEP', 'Burst_Number', 'fmin', 'fmax']].to_csv('../Example_Data/Full_Run_Through/akr_flims.csv', index=False)
```

## Wind Postion
Here we obtain the position of WIND. In the repository we provide the dataset of WIND position and then calculate magnetic local time and magnetic latitude. The WIND position data can be obtained from: https://sscweb.gsfc.nasa.gov/cgi-bin/Locator.cgi. If each time interval is saved as .txt file and the following options are chosen: Choosing options: GSE XYZ, GSE LAT/LON, GSE Local Time, GSM XYZ, GSM LAT/LON, Radial Distance, we provide a function to combine and format the txt files into a new dataframe. Wind_Waves.reading_tools.read_combine_wind_position.

Alternatively the file we provide can be used.

In this step we load the file and apexpy to calculate magnetic local time and latitude. apexpy can be installed with pip: pip install apexpy.

First we calculate magnetic latitude and local time using apex co-ordinates, however due to the "altitude" of WIND apex is not appropriate for magnetic latitude. So the next step we calculate magnetic latitude using quasi-dipole (QD) co-ordinates. More information on this can be found in Laundal and Richmond 2016: https://doi.org/10.1007/s11214-016-0275-y


```python
import pandas as pd
from apexpy import Apex
df= pd.read_csv('../Example_Data/Full_Run_Through/WIND_Position.csv', parse_dates=['Date_UTC'])
RE = 6378100  # Earth's radius in meters
A= Apex(pd.Timestamp('2000')) # We use the year 2000 as it is an approximate midpoint of the dataset. The change per year is small.
df['mlat'], df['mlt']= A.convert(df.GLat, df.GLon, source='geo', dest='mlt', datetime=df.Date_UTC.values, height= ((df.GX_RE**2+df.GY_RE**2+df.GZ_RE**2)**.5 -1)*RE)
df['mlat'], _= A.convert(df.GLat, df.GLon, source='geo', dest='qd', datetime=df.Date_UTC.values, height= ((df.GX_RE**2+df.GY_RE**2+df.GZ_RE**2)**.5 -1)*RE)
df.rename(columns={'radius':'R'}, inplace=True)
df.drop(columns=['x_gse', 'y_gse', 'z_gse', 'lat_gse', 'lon_gse', 'gseLT',
                'x_gsm', 'y_gsm', 'z_gsm', 'lat_gsm', 'lon_gsm',
                'geoLT', 'gseLT_Hr', 'geoLT_Hr']).to_csv('../Example_Data/Full_Run_Through/WIND_Position_Mag.csv', index=False)
```
